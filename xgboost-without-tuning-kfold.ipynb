```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

lb=LabelEncoder()

cat_cols =['region','sex','smoker']

df['region']=lb.fit_transform(df[['region']])
df['sex']=lb.fit_transform(df[['sex']])
df['smoker']=lb.fit_transform(df[['smoker']])

df.head()
```

```python
sts=StandardScaler()

X=df.drop('charges',axis=1)
y=df['charges']


X=sts.fit_transform(X)

X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)

X_train
```

```python
from xgboost import XGBRegressor

from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score


xgb=XGBRegressor()
final_test_predictions = []
scores = []
r2scores = []

kfold = KFold(5)
# enumerate splits
for train, test in kfold.split(X,y):
    #print('train: %s, test: %s' % (X[train], X[test]))
    #print('train: %s, test: %s' % (y[train], y[test]))
    xgb.fit(X[train],y[train])
    y_pred=xgb.predict(X[test])
    rmse = mean_squared_error(y[test], y_pred, squared=False)
    print( rmse)
    scores.append(rmse)
    final_test_predictions.append(y_pred)
    
```
