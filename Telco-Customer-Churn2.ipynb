```python
import numpy as np
import pandas as pd
from sklearn import tree

from sklearn import tree
#import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

```python
df=pd.read_csv("./kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

# Checking Datatypes

```python
df.dtypes
```

# Converting string into float/int

```python
d0 = {'Female': 0, 'Male': 1}
df['gender'] = df['gender'].map(d0)
d = {'No': 0, 'Yes': 1}
df['Partner'] = df['Partner'].map(d)
df['Dependents'] = df['Dependents'].map(d)
df['PhoneService'] = df['PhoneService'].map(d)
d1={'No': 0, 'Yes': 1,'No phone service':2}
df['MultipleLines'] = df['MultipleLines'].map(d1)
df['OnlineSecurity'] = df['OnlineSecurity'].map(d)
```

```python
df.dtypes
```

# Splitting the data

```python
X=df.iloc[:10,1:4] #features
Y=df.iloc[:10,-1] #target/class
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.7, random_state=123) #test size as 30% and random state as 123, random state is for our output can be replicated
```

# Training the algorithm

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y_train_encoded = le.fit_transform(Y_train)

xg_cl=xgb.XGBClassifier(objective='binary:logistic', seed=123) #XGBoost classifier syntax 
xg_cl.fit(X_train, Y_train_encoded) #fix model on X_train and Y_train 
```

# Train Accuracy

```python
xg_cl.score(X_train,Y_train_encoded) #train accuracy on traning set
```

# Test Accuracy

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y_test_encoded = le.fit_transform(Y_test)

xg_cl.score(X_test, Y_test_encoded) #test accuracy on traning set
```

# Prediction

```python
Yp=xg_cl.predict(X_test)
Yp
```
