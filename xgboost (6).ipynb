```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.datasets import load_iris
import random
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import xgboost as xgb

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# Any results you write to the current directory are saved as output.
```

```python
X, y = load_iris(return_X_y=True)
```

```python
Xy = np.column_stack((X,y))
```

```python
### Pomieszanie danych
np.random.shuffle(Xy)
```

```python
### Rozłączenie z powrotem
X, y = Xy[:, :-1], Xy[:, -1]
```

```python
# Podzielenie na zbiór testowy i treningowy
length = int(len(X)* 0.8)
X_train = X[:length]
X_test = X[length:]
y_train = y[:length]
y_test = y[length:]
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
# Setting some parameters

parameters = {
    'eta': 0.3,  
    'silent': True,  # option for logging
    'objective': 'multi:softprob',  # error evaluation for multiclass tasks
    'num_class': 3,  # number of classes to predic
    'max_depth': 3  # depth of the trees in the boosting process
    }  
num_round = 20  # the number of training iterations
```

```python
clfX = xgb.train(parameters, dtrain,num_round)
```

```python
y_predictedX = clfX.predict(dtest)
```

```python
y_predictedX
```

```python
y_test
```

```python
lista =[]
for i in y_predictedX.tolist():
    lista.append(i.index(max(i)))
listN = np.array(lista)
print(listN)
```

```python
clf = LogisticRegression(random_state=0, solver = 'newton-cg') # Definicja modelu
```

```python
### Wytrenowanie modelu
clf.fit(X_train, y_train)
```

```python
### Przewidywanie dla danych testowych
y_predicted = clf.predict(X_test)
```

```python
### Sprawdzenie dokladnosci
y_true  = y_predicted[y_predicted== y_test]
```

```python
y_trueX = listN[listN == y_test]
```
