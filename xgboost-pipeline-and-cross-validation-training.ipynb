```python
df = df.drop('Unnamed: 32', axis = 'columns')
```

```python
df = df.drop('id', axis = 'columns')
```

```python
df.diagnosis = df.diagnosis.map({'B':1, 'M':0})
```

The number of people that in Bening stages is higher than the one  in malignant stages

```python
features_mean = ['radius_mean','texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean',
                'concavity_mean','concave points_mean','symmetry_mean', 'fractal_dimension_mean']
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df[features_mean].corr(), annot=True)
```

```python
predict_feature = ['radius_mean', 'concave points_mean', 'concavity_mean', 'perimeter_mean', 'compactness_mean']
```

```python
x = df[predict_feature]
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

Building the pipeline

```python
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

model = XGBClassifier(n_estimators = 500, learning_rate = 0.05, n_jobs = 4, random_state = 0)
pipe = Pipeline(steps=[('model', model)])
```

```python
pipe.fit(x_train, y_train)
```

```python
preds = pipe.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(pipe, x, y, cv=3, scoring='accuracy')

print(scores)
```
