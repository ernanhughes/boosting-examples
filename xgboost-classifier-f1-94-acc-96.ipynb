```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score
import xgboost as xgb
```

```python
data = pd.read_csv('../input/fetal-health-classification/fetal_health.csv')
```

```python
train_df, test_df = train_test_split(data, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train = train_df.drop(["fetal_health"], axis=1).values
y_train = train_df["fetal_health"].values
```

```python
X_train_std = scaler.fit_transform(X_train)
```

```python
clf = xgb.XGBClassifier()
clf.fit(X_train_std, y_train)
```

```python
X_test = test_df.drop(["fetal_health"], axis=1).values
y_test = test_df["fetal_health"].values
```

```python
X_test_std = scaler.transform(X_test)

y_pred = clf.predict(X_test_std)

print(f1_score(y_test, y_pred, average='macro'))
print(accuracy_score(y_test, y_pred))
```
