```python
import numpy as np
import xgboost
from sklearn.model_selection import train_test_split as tr
from sklearn.metrics import accuracy_score
```

```python
# Load data from numpy
dataset = np.loadtxt('/kaggle/input/pimaindiansdiabetescsv/pima-indians-diabetes.csv', delimiter=",")
```

```python
# Data into x and y
x = dataset[:,0:8]
y = dataset[:,8]
```

```python
# Data train and test sets
seed = 7
test_size = 0.33
x_train, x_test, y_train, y_test = tr(x, y, test_size=test_size, random_state=seed)
```

```python
# Fit model
model = xgboost.XGBClassifier()
model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(x_test)
predictions = [round(value) for value in y_pred]
```

```python
# Accuracy score
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
